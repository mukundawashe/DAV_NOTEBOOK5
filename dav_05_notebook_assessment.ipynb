**Question 4.** Relationships between variables. A scatterplot illustrates a linear relationship between two variables. Consider the  relationship between the dependent variables, `Y1` and `Y2,` and any of the other explanatory variables. Discuss what the relationships you observe. Plot appropriate visualisations and label your plots (title, axes labels, etc) with good descriptions.

```python
plt.figure(figsize=(10,8))
plt.scatter(data['Y1'], data['Y2'], s=20, c='green', marker='o', alpha=.5)
plt.title('Heating Load vs Cooling Load ')
plt.xlabel('HEATING LOAD')
plt.ylabel('COOLING LOAD')
```

```python
The scatter plot shows a strong positive relationship between the cooling load and the heating load. This visual implies that
the cooling load increases as the heating load increases and vice versa.
```

```python
plotting.scatter_matrix(data[['X1','X2', 'Y1', 'Y2']], diagonal='kde')
plt.figure(figsize=(25,8))
```

```python
From the scattermatrix above, it can be inferred that X1 and X2 have a co-relation. Their plot shows a strong negative 
relationship which means that X2 decreases as X1 increases or Surface area increases as Relative Compactness decreases.
Y1 and Y2 also have a strong positive relationship.
```
